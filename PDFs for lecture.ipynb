```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.xkcd()
```

```python
import numpy
import math
```

```python
x = numpy.linspace(-15,15,100)
mu = 4
sigma = 2
y = numpy.exp(-0.5*((x-mu)/sigma)**2)/sigma/math.sqrt(2*math.pi)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
fac = math.factorial
def binomial(N, p):
    def binx(x):
        return fac(N)/(fac(x)*fac(N-x))*(p**x)*(1-p)**(N-x)
    return binx

total = 10
p_ = 0.4
bin_fun = binomial(total, p_)
x = numpy.arange(total + 1)
y = map(bin_fun, x)
plt.plot(x,y)
plt.xlabel('n')
plt.ylabel('f(n)')
plt.annotate(
         'N: {}\np: {}'.format(total,p_),
         xy=(6, 0.2))
plt.show()
```

```python
fac = math.factorial
def poisson(nu):
    expnu = math.exp(-nu)
    def poisx(x):
        return nu**x*expnu/fac(x)
    return poisx

nu = 10
total = 30
p_fun = poisson(nu)
x = numpy.arange(total + 1)
y = map(p_fun, x)
plt.plot(x,y)
plt.xlabel('n')
plt.ylabel('f(n)')
plt.annotate(
         r'$\nu$ : {}'.format(nu),
         xy=(20, 0.08))
plt.show()
```

```python
x = numpy.linspace(0,200,1000)
mu = 80
sigma = 10
y = numpy.exp(-0.5*((x-mu)/sigma)**2)/sigma/math.sqrt(2*math.pi)
plt.plot(x,y)
plt.xlabel('n')
plt.ylabel('f(n)')
plt.annotate(
         r'$\mu$: {}{}$\sigma$: {}'.format(mu,"\n", sigma),
         xy=(120, 0.015))
plt.show()
```
