```python
import matplotlib.pyplot as plt
def show_plt(arr):
    plt.plot(arr)
    plt.show()
```

```python
import numpy as np

# Design our signal
signal = np.zeros(2000)

# Add a jump at 400
signal[400:] += 1

# Add noise
signal += np.random.normal(0, 0.1, len(signal))

show_plt(signal)
```

```python
# Now search for a jump
# The ideal way to do this is to calculate a moving average derivative.  The following does a 25 pt moving average derivative.
mw_avg = np.zeros(50)
mw_avg -= 1
mw_avg[:len(mw_avg)/2] = 1
show_plt(mw_avg)


# With valid, we only convolve where the signal overlap!  (be careful of boundary effects)
new_arr = np.convolve(mw_avg, signal, "valid")

show_plt(new_arr)

# Now we can do a simple threshold filter, find max, compare to avg/min
```
